```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, StratifiedKFold
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Assuming you have defined X, y, categorical_features, and numerical_features elsewhere

def MLpipe_KFold_RMSE(X, y, model, param_grid, cv=5):
    """
    Perform machine learning using KFold cross-validation with a given model.
    
    Parameters:
    - X: Features
    - y: Target variable
    - model: Scikit-learn estimator
    - param_grid: Hyperparameter grid for GridSearchCV
    - cv: Number of folds for cross-validation
    
    Returns:
    - Best hyperparameters and test scores
    """
    # Create a column transformer for encoding
    preprocessor = ColumnTransformer(
        transformers=[
            ('onehot', OneHotEncoder(sparse_output=True, handle_unknown='ignore'), categorical_features),
            ('minmax', MinMaxScaler(), numerical_features)
        ])

    # Create a pipeline with preprocessing and the specified model
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', model)
    ])

    # Use GridSearchCV for hyperparameter tuning
    grid_search = GridSearchCV(pipeline, param_grid, cv=cv, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X, y)

    # Print the best hyperparameters
    print(f"Best hyperparameters for {model.__class__.__name__}:", grid_search.best_params_)

    # Print the performance metrics for each combination of hyperparameters
    cv_results = pd.DataFrame(grid_search.cv_results_)
    print(f"{model.__class__.__name__} CV results:")
    print(cv_results[['params', 'mean_test_score', 'std_test_score']])

    # Perform cross-validation with the best model
    cv_scores = cross_val_score(grid_search.best_estimator_, X, y, cv=cv, scoring='accuracy')
    print(f"Cross-validation Scores for {model.__class__.__name__}:", cv_scores)
    print(f"Mean Accuracy for {model.__class__.__name__}:", cv_scores.mean())

    return grid_search.best_params_, cv_scores

# Linear Regression with L1 Regularization
from sklearn.linear_model import LogisticRegression

# Define the hyperparameter grid for grid search for Logistic Regression
param_grid_logreg = {
    'classifier__C': [100, 10, 1.0, 0.1, 0.01],
    'classifier__penalty': ['l1'],
    'classifier__solver': ['liblinear']
}

# Perform machine learning with Logistic Regression using KFold cross-validation
best_params_logreg, cv_scores_logreg = MLpipe_KFold_RMSE(X, y, LogisticRegression(), param_grid_logreg)

# Print the mean and standard deviation of the test scores
print(f"Mean Test Score for Logistic Regression:", cv_scores_logreg.mean())
print(f"Standard Deviation of Test Scores for Logistic Regression:", cv_scores_logreg.std())
```
